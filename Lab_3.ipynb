# ¿Cuál es el tamaño en bytes de un array de 1000 elementos de tipo booleano?

```python
import sys

array_booleano = [True] * 1000

tamanio_bytes_elemento = sys.getsizeof(True)
tamanio_bytes_total = tamanio_bytes_elemento * len(array_booleano)

print(f"Tamaño del array en bytes: {tamanio_bytes_total}")
```

# Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9], obtenga la media aritmética de la diferencia (a-b).

```python
import numpy as np

a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])

dif = a - b
media_aritmetica = np.mean(dif)

print(f"Media aritmética de la diferencia: {media_aritmetica}")
```

# 5) Obtenga la media, mediana, desviación estándar, valor máximo y valor mínimo de la población de los municipios de la provincia de Granada.

```python
import pandas as pd
df = pd.read_csv("nucleospoblacion.csv")
granada = df[df['Provincia'] == 'Granada']
media= granada['Poblacion'].mean()
mediana = granada['Poblacion'].median()
ds = granada['Poblacion'].std()
maximo = granada['Poblacion'].max()
minimo = granada['Poblacion'].min()
print("Media de la población:", media)
print("Mediana de la población:", mediana)
print("Desviación estándar de la población:", ds)
print("Valor máximo de la población:", maximo)
print("Valor mínimo de la población:", minimo)
```

# 7) Seleccione al azar cincuenta municipios diferentes de entre los diponibles en el archivo. Asegúrese de que no se repitan. ¿Luego calcule el promedio de la población y la desviación estándar de esto 50 municipios?.

```python
import pandas as pd
import random
df = pd.read_csv("nucleospoblacion.csv")
municipios = len(df)
indices_aleatorios = random.sample(range(municipios), 50)
municipios_aleatorios = df.iloc[indices_aleatorios]
promedio_poblacion = municipios_aleatorios['Poblacion'].mean()
dspoblación = municipios_aleatorios['Poblacion'].std()

print("Promedio de los municipios seleccionados al azar:", promedio_poblacion)
print("Desviación estándar de los cincuenta municipios seleccionados al azar:", dspoblación)
```
